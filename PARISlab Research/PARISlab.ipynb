```python
# MANIPULATING DATA INTO OUR VARIABLES

# loads mineral families into a numpy array called mineral_classes_data
# columns will contain AT%*100 of the 9 elements

#array of each mineral class in terms of area percentage from excel sheet
forsterite =  [19.96, 45.48, 0, 34.55, 0, 0 ,0 ,0, 0]
fayalite =    [13.78, 31.40, 0, 0, 0, 0, 54.8, 0 , 0]
anorthite =   [20.86, 46.36, 13.79, 0, 0.42, 0, 0, 0, 18.57]
bytownite =   [22.47, 46.54, 11.66, 0, 0.42, 0, 0, 0, 17.66]
labradorite = [24.8, 47.09, 11.8, 0, 1.69, 0, 0, 0, 15.88]
andesine =    [27.18, 47.65, 8.95, 0, 3.42, 0, 0, 0, 14.06]
oligoclase =  [29.63, 48.22, 6.04, 0, 5.2, 0, 0, 0, 12.2]
albite =      [31.5, 48.66, 3.05, 0, 6.99, 0, 0, 0, 10.77]
quartz =      [46.75, 53.26, 0, 0, 0, 0, 0, 0, 0]
diopside =    [25.95, 44.34, 18.51, 11.23, 0, 0, 0, 0, 0]
hedenbergite = [22.64, 38.69, 16.15, 0, 0, 0, 22.51, 0, 0]
orthoclase =  [30.28, 45.99, 0, 0, 0, 14.05, 0, 0, 9.7]
magnetite =   [0, 27.64, 0, 0, 0, 0, 72.37, 0, 0]
ilmenite =    [0, 31.64, 0, 0, 0, 0, 36.81, 31.55, 0]
enstatite =   [27.97, 47.81, 0, 24.21, 0, 0, 0, 0, 0]
ferrosilite = [21.29, 36.39, 0, 0, 0, 0, 42.34, 0, 0]
none =        [0, 0, 0, 0, 0, 0, 0, 0, 0]

# aggregate all minerals
mineral_classes_data = [forsterite, fayalite, anorthite, bytownite, labradorite, andesine, oligoclase, albite, quartz, diopside, hedenbergite, orthoclase, magnetite, ilmenite, enstatite, ferrosilite, none]

# times by 100 to convert to AT% * 100
for i in range(len(mineral_classes_data)):
    for j in range(len(mineral_classes_data[i])):
        mineral_classes_data[i][j] = mineral_classes_data[i][j] * 100

# turn into numpy
mineral_classes = np.array(mineral_classes_data)

mineral_density = {
    'forsterite': 3.27,
    'fayalite': 4.39,
    'anorthite': 2.73,
    'bytownite': 2.71,
    'labradorite': 2.73,
    'andesine': 2.67,
    'oligoclase': 2.65,
    'albite': 2.62,
    'quartz': 2.65,
    'diopside': 3.28,
    'hedenbergite': 3.56,
    'orthoclase': 2.56,
    'magnetite': 5.17,
    'ilmenite': 4.55,
    'enstatite': 3.2,
    'ferrosilite': 3.88,
    'none': 0.0
}
```

```python
# HELPER FUNCTION FOR DATAFRAME

def multiply_row(row):
    index_value = row.name
    multiplier = mineral_density[index_value]
    return row * multiplier
```

```python
# MANIPULATING DATA FOR CLASSIFICATION

target_labels = ['forsterite', 'fayalite', 'anorthite', 'bytownite', 'labradorite', 'andesine', 'oligoclase', 'albite', 'quartz', 'diopside', 'hedenbergite', 'orthoclase', 'magnetite', 'ilmenite', 'enstatite', 'ferrosilite', 'none']

X_train = mineral_classes
y_train = np.array(target_labels)

X_test = samples.to_numpy()
```

```python
# MODEL FOR CLASSIFICATION AND PREDICTING

# we train our model using the mineral classes
# then we make predictions on our sample data

from sklearn.neighbors import KNeighborsClassifier

# Create a KNeighborsClassifier object
knn = KNeighborsClassifier(n_neighbors=1)

# Fit the classifier to the training data
knn.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn.predict(X_test)

# Print predictions dataframe
y_pred_df = pd.DataFrame(y_pred, columns = ['mineral'])
display(y_pred_df)
```

```python
# DATA RESULTS

from collections import Counter
c = Counter(y_pred)

c_df = pd.DataFrame.from_dict(c, orient='index', columns = ['count'])

print(c_df)

#multiply_row was implemented above in helper functions

c_df = c_df.apply(multiply_row, axis=1)
c_df = c_df.rename(columns={'count':'Mass'})
print(c_df)
```

```python
mineral_families = {}
mineral_families['olivine'] = set(['forsterite','fayalite'])
mineral_families['plagioclase'] = set(['anorthite','bytownite','labradorite','andesine','oligoclase','albite'])
mineral_families['pyroxene'] = set(['diopside', 'hedenbergite', 'enstatite', 'ferrosilite'])

oliv_map_dict = {val: num for num, val in enumerate(['forsterite','fayalite'])}
plag_map_dict = {val: num for num, val in enumerate(['anorthite','bytownite','labradorite','andesine','oligoclase','albite'])}
pyro_map_dict = {val: num for num, val in enumerate(['diopside', 'hedenbergite', 'enstatite', 'ferrosilite'])}

plag_encoded_pred = []
oliv_encoded_pred = []
pyro_encoded_pred = []

for mineral in y_pred:
  if mineral not in mineral_families['olivine']:
    oliv_encoded_pred.append(2)
  else:
    oliv_encoded_pred.append(oliv_map_dict[mineral])

  if mineral not in mineral_families['plagioclase']:
    plag_encoded_pred.append(6)
  else:
    plag_encoded_pred.append(plag_map_dict[mineral])


  if mineral not in mineral_families['pyroxene']:
    pyro_encoded_pred.append(4)
  else:
    pyro_encoded_pred.append(pyro_map_dict[mineral])


numpy_plag_encoded_preds = np.array(plag_encoded_pred)

reshaped_plag_encoded_preds = numpy_plag_encoded_preds.reshape(220,256)

numpy_oliv_encoded_preds = np.array(oliv_encoded_pred)

reshaped_oliv_encoded_preds = numpy_oliv_encoded_preds.reshape(220,256)

numpy_pyro_encoded_preds = np.array(pyro_encoded_pred)

reshaped_pyro_encoded_preds = numpy_pyro_encoded_preds.reshape(220,256)
```
